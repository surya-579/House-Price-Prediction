# **Seperating Features-Labels and Splitting Train-Test portions**

```python
from sklearn.utils import shuffle
df1 = shuffle(df)
```

```python
df.insert(9, "median_house_value", df.pop("median_house_value"))
df.head()
```

Lets split the data into two poritons:


*   X - Features
*   Y - Labels



```python
data = df1.values
X = data[: , 1:10]
Y = data[: , 10]
print(X, Y)
```

Now split these into 4 sections:


*   x_train - features of the training data
*   y_train - labels of the training data
*   x_test - features of the testing data
*   y_test - labels of the testing data





```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

*test_train_split* splits the given two compnents X,Y (features, labels) into 4 parts in the ratio of "test_size" i.e;

Here test_size = 0.2, means data taken for testing is 20% of the total available data.

So, 80% of the data is given for training and 20% for testing.



# **Implementing SVC Model from sklearn**

Let use the SVC model from the support vector machine in the sklearn library.

```python
from sklearn.svm import SVC

model1 = SVC()
model1.fit(x_train, y_train)
```

Here *fit()* actually trains the model using the training data.

```python
pred1 = model1.predict(x_test)
```

Here *predict()* is used to make some predictions on the testing data, using the trained model.

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, pred1)
print(accuracy)
```

# **Implementing Logistic Regression Model from sklearn**

```python
from sklearn.linear_model import LogisticRegression

model2 = LogisticRegression()
model2.fit(x_train, y_train)
```

```python
pred = model2.predict(x_test)

accuracy = accuracy_score(y_test, pred)
print(accuracy)
```
